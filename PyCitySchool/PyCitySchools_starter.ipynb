```python
# Program written by Radha Mahalingam. 3-11-19
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)

student_data = pd.read_csv(student_data_to_load)


#Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# This is the code to bring out above 'District Summary'

school_data_complete["budget"] = pd.to_numeric(school_data_complete["budget"])
school_data_complete["math_score"] = pd.to_numeric(school_data_complete["math_score"])
school_data_complete["reading_score"] = pd.to_numeric(school_data_complete["reading_score"])

Tot_sch = school_data_complete['school_name'].nunique()
Tot_std = school_data_complete['student_name'].count()
Tot_Bgt = school_data['budget'].sum()
Avg_mscore = school_data_complete['math_score'].mean()
Avg_rscore = school_data_complete['reading_score'].mean()
Per_pmath = school_data_complete[school_data_complete['math_score'] >= 70].count()['student_name']
Per_pmath = (Per_pmath / Tot_std)*100
Per_pread = school_data_complete[school_data_complete['reading_score'] >= 70].count()['student_name']
Per_pread = (Per_pread / Tot_std)*100
Per_oscore = (Avg_mscore + Avg_rscore) / 2

columns=['Total School','Total Student','Total Budget','Average Math Score','Average Reading Score',
          '% Passing Math','% Passing Reading', 'Overall Passing Score']
District_summary = pd.DataFrame({'Total School': ["{:,}".format(Tot_sch)],
                                 'Total Student': ["{:,}".format(Tot_std)], 
                                 'Total Budget': ["${:,.2f}".format(Tot_Bgt)],
                                 'Average Math Score': ["{:.2f}".format(Avg_mscore)], 
                                 'Average Reading Score': ["{:.2f}".format(Avg_rscore)],
                                  '% Passing Math' : ["{0:.2f}".format(Per_pmath)],
                                 '% Passing Reading': ["{0:.2f}".format(Per_pread)],
                                'Overall Passing Score': ["{0:.2f}".format(Per_oscore)]}, columns = columns)

blankIndex=[''] * len(District_summary)
District_summary.index=blankIndex
District_summary
```

```python
# This is the code to bring out above 'School Summary'

sgb_df = school_data_complete.groupby(['school_name'])

Tot_std = sgb_df['student_name'].count()
Tot_Bgt = sgb_df['budget'].first()
Sch_type = sgb_df['type'].first()
Sch_name = sgb_df['school_name'].first()
Per_Std_Bgt = (Tot_Bgt/Tot_std)
Avg_mscore = sgb_df['math_score'].mean()
Avg_rscore = sgb_df['reading_score'].mean()
Per_pmath=school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name')['math_score'].count()
Per_pmath = (Per_pmath / Tot_std)*100                                                                               
Per_pread=school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name')['reading_score'].count()
Per_pread = (Per_pread / Tot_std)*100 
Per_oscore = (Per_pmath + Per_pread) / 2

School_summary = pd.DataFrame({'school_name' : Sch_name,
                               'School Type' : Sch_type,
                               'Total Student': Tot_std.map('{:,}'.format),
                               'Total School Budget': Tot_Bgt.map('${:,.2f}'.format),
                               'Per Student Budget' : Per_Std_Bgt.map('{:,.2f}'.format),
                               'Average Math Score' : Avg_mscore.map('{:.2f}'.format),
                               'Average Reading Score' : Avg_rscore.map('{:.2f}'.format),
                               '% Passing Math' : Per_pmath.map('{:.2f}'.format),
                               '% Passing Reading' : Per_pread.map('{:.2f}'.format),
                               'Overall Passing Rate' : Per_oscore.map('{:.2f}'.format)})

#School_summary = School_summary.reset_index(drop=True)

School_summary.head() 
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# This is the code to bring out 'Scores by School Spending'

spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]


School_summary["Per Student Budget"] = pd.to_numeric(School_summary["Per Student Budget"])
School_summary["Average Math Score"] = pd.to_numeric(School_summary["Average Math Score"])
School_summary["Average Reading Score"] = pd.to_numeric(School_summary["Average Reading Score"])
School_summary["% Passing Math"] = pd.to_numeric(School_summary["% Passing Math"])
School_summary["% Passing Reading"] = pd.to_numeric(School_summary["% Passing Reading"])
School_summary["Overall Passing Rate"] = pd.to_numeric(School_summary["Overall Passing Rate"])


group_spend = School_summary.groupby(pd.cut(School_summary['Per Student Budget'], 
                                        bins=spending_bins, labels=group_names))

avg_math_score = group_spend["Average Math Score"].mean()
avg_reading_score = group_spend["Average Reading Score"].mean()
pct_pass_math = group_spend["% Passing Math"].mean()
pct_pass_reading = group_spend["% Passing Reading"].mean()
pct_pass_overall = group_spend["Overall Passing Rate"].mean()

school_spend_df = pd.DataFrame({"avg_math_score": avg_math_score.map('{:.2f}'.format),
                              "avg_reading_score": avg_reading_score.map('{:.2f}'.format),
                               "pct_pass_math": pct_pass_math.map('{:.2f}'.format),
                                "pct_pass_reading": pct_pass_reading.map('{:.2f}'.format),
                                "pct_pass_overall": pct_pass_overall.map('{:.2f}'.format)
                               }
                            )
school_spend_df
```
